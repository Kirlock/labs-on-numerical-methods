```python
from math import sin, cos, exp, sqrt, pi
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
f = lambda x,y: -np.cos(x) * np.cos(y) + 1

x = np.arange(-3 * pi, 3 * pi + 0.2, 0.2)
y = np.arange(-2 * pi, 2 * pi + 0.2, 0.2).reshape(-1, 1)
z = f(x, y)

plt.contourf(x, y.T[0], z)

h = 0.2
for ind in range(1, 40):
    x0 = -7.5 + 0.4 * (ind - 1)
    y0 = -6 + 0.1 * (ind - 1)
    for i in np.arange(1, 90):
        x1 = x0 - f(x0 + h/2, y0) + f(x0 - h/2, y0)
        y1 = y0 - f(x0, y0 + h/2) + f(x0, y0 - h/2)
        plt.plot([x0, x1], [y0, y1])
        x0 = x1
        y0 = y1
        
plt.show()
```
