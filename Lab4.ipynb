```python
import math
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    l = math.pi / 2
    if x >= 0 and x < l/3:
        return x*3/l
    else:
        return 1
    
n=70
xl=math.pi/2
a=1
x = np.arange(0, math.pi, 1 / 640)
jr = np.arange(1, 20, 1)
wn = [0]*100

points_initial1 = []

for i in range(1, n-1):
    wn[i]=0
    for j in range(1, n-1):
        x1=xl/(n-1)*j
        wn[i]=wn[i]+(f(x1)-x1/xl)*math.sin(math.pi*i*j/(n-1))*2/(n-1)
        
for il in x:
    points_initial1.append([il, f(il)])
x_func = [p[0] for p in points_initial1]
y_func = [p[1] for p in points_initial1]
plt.plot(x_func, y_func, label = "initial")

fourier = []

for j in jr:
    fourier = []
    t = xl**2/a**2/200*(j-1)**2
    for il in range(1,640):
        x1 = il / 639 * xl
        s = x1 / xl
        for i in range(1, n-1):
            s2=1
            s1=math.exp(-i*t)
            for j1 in range(1, i+1):
                s2=s2*s1
            s+=wn[i]*s2*math.sin(math.pi*i*x1/xl)
        fourier.append([x1,s])
    x_func = [p[0] for p in fourier]
    y_func = [p[1] for p in fourier]
    plt.plot(x_func, y_func, label = "j = {}".format(j))


plt.rcParams["figure.figsize"] = (10,10)
plt.grid(True)
plt.legend()
plt.ylim([0, 1.2])
plt.xlim([0, 1.6])
plt.show()
    
```
