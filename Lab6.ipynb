```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

f = lambda x,y: 3*x*x/(x**3+y+1)
f1 = lambda x: -x**3-2
x = x0 = 0
y0 = y21 =  y11 = y3 = y41 = -2
n = 20
x1 = 1
h1 = x1/n

x_f1 = np.linspace(0, 1, n)
f1_val = [f1(x) for x in x_f1]
y2_val = []
y1_val = []
y4_val = []
y3_val = []

plt.plot(x_f1, f1_val, label = "f1")
plt.title("Accurate sol.")
plt.xlim([0, 1])
plt.ylim([-2.82, -1.99])
plt.show()

for x1 in x_f1:
    h = x1-x
    if h > h1:
        y21 = y21 + f(x,y21)*h1
        s = f(x,y11)
        y11 = y11+f(x+h1/2,y11+h1/2*s)*h1
        
        k1 = f(x,y41)
        k2 = f(x+h1/2,y41+h1/2*k1)
        k3 = f(x+h1/2,y41+h1/2*k2)
        k4 = f(x+h1/2,y41+h1/2*k3)
        y41 = y41+(k1+2*k2+2*k3+k4)*h1/6
        x = x+h1
        h = x1-x
        
    y1_val.append(y21)
    y2_val.append(y11)
    y4_val.append(y41)
        
plt.plot(x_f1, y1_val)
plt.title("Euler sol.")
plt.xlim([0, 1])
plt.ylim([-2.82, -1.99])
plt.show()

plt.plot(x_f1, y2_val)
plt.title("2nd order sol.")
plt.xlim([0, 1])
plt.ylim([-2.82, -1.99])
plt.show()

#plt.plot(x_f1, y3_val)
#plt.title("2nd order_2 sol.")
#plt.show()

plt.plot(x_f1, y4_val)
plt.xlim([0, 1])
plt.ylim([-2.82, -1.99])
plt.title("4th order sol.")
plt.show()

plt.plot(x_f1, f1_val, label = "f1")
plt.plot(x_f1, y1_val, label = "Euler")
plt.plot(x_f1, y2_val, label = "2nd order")
#plt.plot(x_f, y3_val, label = "2nd order_2")
plt.plot(x_f1, y4_val, label = "4th order")
plt.title("Aggregate")
plt.legend()
plt.xlim([0, 1])
plt.ylim([-2.82, -1.99])
plt.show()
```
