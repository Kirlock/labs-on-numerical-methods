```python
import math as m
import matplotlib.pyplot as plt
```

```python
f = lambda x : m.sin(x)*m.sqrt(x)+1
calc_y_point = lambda x : int(48)
```

```python
n = 5
xl = 2 * m.pi
xln = xl / 2
h = m.pi / 7
mult = [0.5, 1.5, 2.5, 4.5, 6.5]
X = [h * i for i in mult]
y10 = m.pi
y0 = 2
```

```python
points_base = []
points_Lagrange = []
points_func = []
splitting = 1000
for i in range(1, splitting + 1):
    x1 = i / splitting * xl
    x10 = i / splitting * xln
    s1 = 0
    for ii in range(1, n + 1):
        s = 1
        s10 = 1
        for jj in range(1, n + 1):
            if (jj != ii):
                s *= (x1-X[jj - 1])/(X[ii - 1] - X[jj - 1])
                s10 *= (x10-X[jj - 1])/(X[ii - 1] - X[jj - 1])
        s1 += s * f(X[ii - 1])
        points_base.append([x10, s10])
    points_Lagrange.append([x1, s1])
    points_func.append([x1, f(x1)])
```

```python
x_L = [p[0] for p in points_Lagrange]
y_L = [p[1] for p in points_Lagrange]
x_func = [p[0] for p in points_func]
y_func = [p[1] for p in points_func]

plt.plot(x_func, y_func, label = "f(x)", color = "r")
plt.plot(x_L, y_L, label = "Lagrange")
for point in X:
    plt.scatter(point, f(point))
    
x_graph = [p[0] for p in points_base]
y_graph = [p[1] for p in points_base]
plt.scatter(x_graph , y_graph, s = 5, label = "base functions")

plt.rcParams["figure.figsize"] = (10,10)
plt.grid(True)
plt.legend()
plt.savefig("lagrange.png")
plt.show()
```
